# Scatter Plot(MATLAB Method)

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# plt.plot()
plt.plot(x_axis, y_axis, 'o')
```

```python
# plt.scatter()
plt.scatter(x_axis, y_axis)
```

```python
# Create the plot.
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
plt.xlim(0,50)
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

# Scatter Plot and bubble char(Object-oriented-interface)

```python
# Scatter Plot(Object-oriented-interface)
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create the plot.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, marker=dict(color="rebeccapurple", size=20, line=dict(width=2,color='MediumPurple'), label='Boston')
ax.invert_yaxis()
ax.set_xlim(0,50)
# Create labels for the x and y axes.
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.legend()

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*5)

ax.scatter(y_axis, x_axis, s=y_axis_larger)
```

# Pie Chart (Matlab Method)

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

# Pie Chart (object-oriented interface)

```python
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow=True,
    counterclock=False,
    startangle=95)

plt.show()
```
